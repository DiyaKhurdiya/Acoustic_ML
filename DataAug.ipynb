Adding Noise using the GaussianNoise function from Audiomentations

```python
def add_white_noise(signal, noisefactor):
  noise = np.random.normal(0, signal.std(), signal.size)
  augmentedsignal = signal + noise*noisefactor
  return augmentedsignal
```

```python
aug_signal = add_white_noise(x1, 0.1)
aug_signal1 = add_white_noise(x2, 0.5)
aug_signal2 = add_white_noise(x2, 1)
aug_signal3 = add_white_noise(x2, 1.5)
aug_signal4 = add_white_noise(x2, 3)
aug_signal5 = add_white_noise(x2, 5)


```
